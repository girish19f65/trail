```python
# Student Grade Manager

students = [("King", 90), ("Queen", 80), ("Jack", 70)]

while True:
    print("\n STUDENT GRADE MANAGER")
    print("1. Add Student")
    print("2. View Students")
    print("3. Update Grade")
    print("4. Delete Student")
    print("5. Exit")

    choice = input("Enter your choice: ")
#Add Student
    if choice == '1':
        name = input("Enter student name: ").capitalize()
        grade = float(input("Enter grade: "))
        student = (name, grade)
        students.append(student)
        print("Student added successfully!")
#View All Students
    elif choice == '2':
        if not students:
            print("No students found!")
        else:
            print("\n--- Student List ---")
            i = 1
            for student in students:
                name, grade = student
                print(f"{i}. Name: {name}, Grade: {grade}")
                i += 1
#Update a Student's Grade
    elif choice == '3':
        if not students:
            print("No students to update!")
        else:
            update_name = input("Enter student name to update: ").capitalize()
            found = False
            for i in range(len(students)):
                if students[i][0] == update_name:
                    new_grade = float(input(f"Enter new grade for {update_name}: "))
                    students[i] = (update_name, new_grade)
                    print(f"Grade updated for {update_name}!")
                    found = True
                    break
            if not found:
                print("Student not found!")
#Delete a Student
    elif choice == '4':
        if not students:
            print("No students to delete!")
        else:
            del_name = input("Enter student name to delete: ").capitalize()
            found = False
            for student in students:
                if student[0] == del_name:
                    students.remove(student)
                    print(f"{del_name} removed successfully!")
                    found = True
                    break
            if not found:
                print(" Student not found!")
# Exit
    elif choice == '5':
        print(" Thank you for using the Student Grade Manager!")
        break

    else:
        print(" Invalid choice! Please enter a number from 1 to 5.")
```
