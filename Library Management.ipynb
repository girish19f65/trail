```python
# Personal Library Management System

library = {}
while True:
    print("\n1.Add 2.Remove 3.Search 4.Update 5.View Library 6.Statistics 7.Exit ")
    ch = input("Choice: ")

    # ADD
    if ch == "1":
        title = input("Enter title: ")
        author = input("Enter author: ")
        year = input("Enter year: ")
        pages = input("Enter pages: ")

        library[title] = {
            "author": author,
            "year": year,
            "pages": pages
        }
        print("Book added\n")

    # REMOVE
    elif ch == "2":
        title = input("Enter title to remove: ")
        if title in library:
            library.pop(title)
            print("Book removed\n")
        else:
            print("Book not found\n")

    # SEARCH
    elif ch == "3":
        title = input("Enter title to search: ")
        if title in library:
            print(title, ":", library[title])
        else:
            print("Book not found.\n")

    # UPDATE
    elif ch == "4":
        title = input("Enter title to update: ")
        if title in library:
            field = input("Update (author/year/pages): ")
            if field in library[title]:
                new_value = input("Enter new value/authorname: ")
                library[title][field] = new_value
                print("Updated\n")
            else:
                print("Invalid field!\n")
        else:
            print("Book not found!\n")

    # VIEW LIBRARY
    elif ch == "5":
        print("\nLibrary Collection:")
        if not library:
            print("Library is empty.\n")
        else:
            for t, info in library.items():
                print(t, ":", info)

    # STATISTICS
    elif ch == "6":
        if not library:
            print("No statistics. Library empty.\n")
        else:
            print("\n--- Statistics ---")

            # Total books
            print("Total Books:", len(library))

            # Total pages
            total_pages = sum(
                int(info["pages"]) for info in library.values() if info["pages"].isdigit())
            print("Total Pages:", total_pages)

            # Books per author
            author_count = {}
            for info in library.values():
                a = info["author"]
                author_count[a] = author_count.get(a, 0) + 1
            print("Books per Author:", author_count)

            # Oldest & Newest books
            years = {t: int(info["year"])
                     for t, info in library.items() if info["year"].isdigit()}
            oldest = min(years, key=years.get)
            newest = max(years, key=years.get)

            print("Oldest Book :", oldest, "(", years[oldest], ")")
            print("Newest Book :", newest, "(", years[newest], ")\n")

    # EXIT
    elif ch == "7":
        print("Thank you for using the library system.")
        break

    else:
        print("Invalid choice. Choose between 1â€“7.\n")
```
