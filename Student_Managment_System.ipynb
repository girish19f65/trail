```python
# Student Management System using Dictionary of Sets

courses = {
    "Maths": set(),
    "Science": set(),
    "English": set(),
    "Computer": set()
}

while True:
    print("\n1. Enroll Student  2. View Course Students  3. View All Students And Their Courses  4. Exit")
    ch = input("Enter choice: ").strip()

    # Enroll a student in a course
    if ch == "1":
        name = str(input("Enter Student Name: "))
        if not name:
            print("Name cannot be empty. Try again.")
            continue

        print("Available courses:", ", ".join(sorted(courses.keys())))
        course = str(input("Enter Course Name: "))
        if not course:
            print("Course name cannot be empty.")
            continue
        if course not in courses:
            courses[course] = set()
            print(
                f"Course '{course}' did not exist and was created automatically.")

        courses[course].add(name)
        print(f"{name} enrolled in {course}.")

    # View students in a courses
    elif ch == "2":
        course = str(input("Enter Course Name: "))
        if course in courses:
            students = courses[course]
            if students:
                print(f"Students in {course}:")
                for s in sorted(students):
                    print(" -", s)
            else:
                print(f"No students enrolled in {course} yet.")
        else:
            print(f"Course '{course}' not found.")

    # View all courses with students
    elif ch == "3":
        print("\n--- All Courses and Students ---")
        for c in sorted(courses.keys()):
            students = courses[c]
            print(
                f"{c}: {', '.join(sorted(students)) if students else '(no students)'}")

    # Exit from the program
    elif ch == "4":
        print(" Exiting...Thank you")
        break

    else:
        print("Invalid choice â€” please choose a number from 1 to 41.")
```
