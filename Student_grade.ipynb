```python
# student grade calculator
grade_scale = {
    "A": 90,
    "B": 80,
    "C": 70,
    "D": 60,
    "F": 50
}


def calculate_percentage(marks_obtained, total_marks):
    """
    Calculating percentage of a student.
    """
    percentage = (marks_obtained / total_marks) * 100
    return percentage


def get_grade(percentage):
    """
    Grade of a student based on global grade scale.
    """
    global grade_scale

    if percentage >= grade_scale["A"]:
        return "A"
    elif percentage >= grade_scale["B"]:
        return "B"
    elif percentage >= grade_scale["C"]:
        return "C"
    elif percentage >= grade_scale["D"]:
        return "D"
    elif percentage <= grade_scale["F"]:
        return "F"


def main():
    """
    Take marks and prints percentage & grade of a student.
    """
    print("\n--- DOCSTRINGS ---")
    print(calculate_percentage.__doc__)
    print(get_grade.__doc__)
    print(main.__doc__)

    name = input("Enter student name: ")
    marks = float(input("Enter marks obtained: "))
    total = float(input("Enter total marks: "))

    percent = calculate_percentage(marks, total)
    grade = get_grade(percent)

    print("\n--- Result ---")
    print("Student Name:", name)
    print("Percentage: {:.2f}%".format(percent))
    print("Grade:", grade)

main()
```
