```python
# To-Do List Manager

List = []
def show_menu():
    print("\n TO-DO LIST MANAGER ")
    print("1. Add Task")
    print("2. View Tasks List")
    print("3. Exit")

def add_task():
    task = input("Enter a new task: ")
    List.append(task)
    print("Task added successfully.")

def view_tasks():
    print("\n--- Your Tasks ---")
    if len(List) == 0:
        print("No tasks available.")
    else:
        c = 0
        for t in List:
            c += 1
            print(c, ".", t)

while True:
    show_menu()
    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        add_task()

    elif choice == "2":
        view_tasks()

    elif choice == "3":
        # Exit the program
        print("Thank you")
        break

    else:
        print("Invalid choice.")
```
